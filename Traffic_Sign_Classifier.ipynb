### Pre-process the Data Set (normalization, grayscale, etc.)

Minimally, the image data should be normalized so that the data has mean zero and equal variance. For image data, `(pixel - 128)/ 128` is a quick way to approximately normalize the data and can be used in this project. 

Other pre-processing steps are optional. You can try different techniques to see if it improves performance. 

Use the code cell (or multiple code cells, if necessary) to implement the first step of your project.

````python
### Preprocess the data here. It is required to normalize the data. Other preprocessing steps could include 
### converting to grayscale, etc.
### Feel free to use as many code cells as needed.
import skimage.exposure
import tensorflow as tf

def grayscale(images):
    # convert the rgb images to grayscale
    return tf.Session().run(tf.image.rgb_to_grayscale(images))

def normalize(images):
    # normalize the images so pixel values are between -1 to 1
    return (images.astype(np.int) - 128) / 128

def sharpen(images):
    # increate the contrast of the image using histogram equalization method
    return skimage.exposure.equalize_hist(images)

def preprocess_dataset(dataset):
    return sharpen(normalize(grayscale(dataset)))

# examine some examples and the effects of preprocessing
fig = plt.figure(figsize=(14, 14))

a = fig.add_subplot(141)
a.set_title("Original", fontsize=20)
plt.imshow(X_train[5463])

b = fig.add_subplot(142)
b.set_title("Grayscale", fontsize=20)
plt.imshow(grayscale(X_train[5463]).squeeze(), cmap='gray')

c = fig.add_subplot(143)
c.set_title("\n".join(wrap("Grayscale + Normalize", 20)), fontsize=20)
plt.imshow(normalize(grayscale(X_train[5463])).squeeze(), cmap='gray')

d = fig.add_subplot(144)
d.set_title("\n".join(wrap("Grayscale + Normalize + Sharpen", 20)), fontsize=20)
plt.imshow(preprocess_dataset(X_train[5463]).squeeze(), cmap='gray')
````

````python
# preprocess all the dataset
X_train_normalized = preprocess_dataset(X_train)
X_valid_normalized = preprocess_dataset(X_valid)
X_test_normalized = preprocess_dataset(X_test)

# look at the data that was normalized, and make sure the pixel value range after normalization is -1 to 1
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.hist(X_valid.flatten())
ax1.set_title('Original Pixel Values')
ax2.hist(X_valid_normalized.flatten())
ax2.set_title('Preprocessed Pixel Values')
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.
from tensorflow.contrib.layers import flatten

MU = 0
SIGMA = 0.1

def add_conv_layer(inp, filter_2d_size, num_filters, stride, padding):
    inp_channels = inp.get_shape()[-1].value
    filter_shape = filter_2d_size + [inp_channels] + [num_filters]
    conv_w = tf.Variable(tf.truncated_normal(shape=filter_shape, mean=MU, stddev=SIGMA))
    conv_b = tf.Variable(tf.zeros([num_filters]))
    conv = tf.nn.conv2d(inp, conv_w, strides=[1, stride, stride, 1], padding=padding) + conv_b
    
    return tf.nn.relu(conv)

def add_fc_layer(inp, output_size, with_activation=True):
    inp_len = inp.get_shape()[1].value   # the first dimension is batch size, which is None
    fc_w = tf.Variable(tf.truncated_normal(shape=[inp_len, output_size], mean=MU, stddev=SIGMA))
    fc_b = tf.Variable(tf.zeros([output_size]))
    logits = tf.add(tf.matmul(inp, fc_w), fc_b)

    return tf.nn.relu(logits) if with_activation else logits


def TrafficSignClassifer(x, dropout_prob):
    # Layer 1: convolutional layer with 32 x 32 x 3, output is 32 x 32 x 10
    conv1 = add_conv_layer(x, filter_2d_size=[5, 5], num_filters=10, stride=1, padding='SAME')
    
    # Pooling layer: from 32 x 32 x 10 to 16 x 16 x 10
    conv1 = tf.nn.max_pool(conv1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # Layer 2: convolutional layer with 16 x 16 x 10, output is 16 x 16 x 20
    conv2 = add_conv_layer(conv1, filter_2d_size=[5, 5], num_filters=20, stride=1, padding='SAME')
    
    # Pooling layer: from 16 x 16 x 20 to 8 x 8 x 20
    conv2 = tf.nn.max_pool(conv2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    # flatten the conv2 neurons from 3D to 1D
    conv2_flattened = flatten(conv2)
    
    # Fully connected network 1: conv2 has 1280 neurons, connect to about 200 neurons in next layer
    fc1 = add_fc_layer(conv2_flattened, 200)
    fc1 = tf.nn.dropout(fc1, dropout_prob)
    
    # Fully connected network 1: fc1 has 200 neurons, connect to about 100 neurons in next layer
    fc2 = add_fc_layer(fc1, 100)
    fc2 = tf.nn.dropout(fc2, dropout_prob)
    
    # Fully connected network 1: fc2 has 100 neurons, connect to about 43 neurons in next layer
    return add_fc_layer(fc2, n_classes, with_activation=False)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

image_shape = X_train_normalized[0].shape

x = tf.placeholder(tf.float32, (None, *image_shape))
y = tf.placeholder(tf.int32, (None))
one_hot_y = tf.one_hot(y, n_classes)

keep_prob = tf.placeholder(tf.float32)
````

````python
# define the model
logits = TrafficSignClassifer(x, keep_prob)

# define how to compute the error term for indivial output-label pair
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)

# do a mean operation on the individual errors as the loss function
loss_func = tf.reduce_mean(cross_entropy)

# define an optimizer and feed in a handle of the loss function
learn_rate = 0.001
optimizer = tf.train.AdamOptimizer(learning_rate=learn_rate)
training_operation = optimizer.minimize(loss_func)
````

````python
# get a list of array with true (correct prediction) and false (incorrect prediction) values
correction_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
# get the fraction of times that the prediction is correct
accuracy_operation = tf.reduce_mean(tf.cast(correction_prediction, tf.float32))

BATCH_SIZE = 128

def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob:1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples
    
````

````python
from sklearn.utils import shuffle
# actually training a model

EPOCHS = 30
saver = tf.train.Saver()
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)

    print("Starting to train model ...")
    for i in range(EPOCHS):
        X_train_normalized, y_train = shuffle(X_train_normalized, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            x_batch, y_batch = X_train_normalized[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x:x_batch, y:y_batch, keep_prob:0.5})

        training_accuracy = evaluate(X_train_normalized, y_train)
        validation_accuracy = evaluate(X_valid_normalized, y_valid)
        print("Epoch {}:".format(i+1))
        print("    Training Accuracy is: {:.3%}".format(training_accuracy))
        print("    Validation Accuracy is: {:.3%}".format(validation_accuracy))
    
    saver.save(sess, "./traffic_sign_classifier" )
    print("Session saved!")
````
